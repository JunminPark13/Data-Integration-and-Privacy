## 공연장 시설정보 Crawling

```python
number=[i for i in range(5000)]
num=[]

for i in range(len(number)):
    if len(str(number[i]))==1:
        num.append(str('000')+str(i))
    elif len(str(number[i]))==2:
        num.append(str('00')+str(i))
    elif len(str(number[i]))==3:  
        num.append(str('0')+str(i))
    elif len(str(number[i]))==4:  
        num.append(str(i))
```

```python
crawled_data = []

for i in num:
    if int(i) % 300 == 0 : print('잠시대기')
        
    try:
        url = 'http://www.kopis.or.kr/por/db/prfplc/prfplcView.do?menuId=MNU_00021&mt10Id=FC00' + i
        req = requests.get(url)
        html = req.text
        soup = BeautifulSoup(html, 'html')
        name = soup.select('dd')
        table = soup.select('div.sgl div')
        a = re.compile('[가-힣]+').findall(str(table))
        if '편의시설' in a : b = 1
        else : b = 0
        if '장애인시설' in a : c = 1
        else : c = 0
        if '주차시설' in a : d = 1
        else : d = 0
    except :
        sisul = 'None'
        
    if name[0].text != '' :
        crawled_data.append([name[0].text, b,c,d])
    
data = pd.DataFrame(crawled_data, columns=['name','pyuneui','jangaein','jucha'])
data.head()
```

## 공연장 정보

```python
df = pd.read_csv('C:/Users/Junmin/Downloads/theaterDB.csv', encoding='euc-kr')
df.head()
```

```python
df.columns = ['no.','name','n_theater','n_seats','theater_name','n_seats2','attr','open','city','gu','address']
df.head()
```

```python
df = df[['name','n_theater','n_seats','theater_name','n_seats2','attr','open','city','gu','address']]
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df2 = pd.DataFrame(index=range(0,1000),columns=['name', 'n_theater', 'n_seats', 'theater_name', 'n_seats2', 'attr',
       'open', 'city', 'gu', 'address'])
df2.head()
```

```python
a = 0
index = []

for i in range(len(df)):
    if '\n' in df['theater_name'][i]:
        b = df['theater_name'][i].split('\n')
        c = df['n_seats2'][i].split('\n')
        for j in range(len(b)):
            df2.iloc[a+j,:] = df.iloc[i,:]
            df2['theater_name'][a+j] = b[j]
            df2['n_seats2'][a+j] = int(c[j].replace(',',''))
        a = a + len(b)
        index.append(i)

df2.head()
```

```python
df2.shape
```

```python
df2 = df2.dropna(axis=0)
df2.shape
```

```python
df1 = df.copy()
df1 = df1.drop(index)
df1.shape
```

```python
df1.index = range(len(df1))
df1.head()
```

```python
df_final = pd.concat([df1, df2])
df_final.head()
```

```python
df_final.index= range(len(df_final))
df_final.tail()
```

```python
df_final.shape
```

```python
df_final.to_csv('theater_final.csv', encoding='euc-kr', index=False)
```
