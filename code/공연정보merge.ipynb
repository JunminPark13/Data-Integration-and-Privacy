```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

```python
show = pd.read_csv('2018,2019 공연.csv')
show.head()
```

```python
show.columns = ['name','genre','start','end','place','theater','seats','company']
show.head()
```

```python
theater = pd.read_csv('final_merge.csv')
theater.head()
```

```python
theater = theater[['name', 'n_theater', 'n_seats', 'theater_name',
       'n_seats2', 'attr', 'open', 'city', 'gu', 'address', 'CS2', 'PK6',
       'OL7', 'SW8', 'CT1', 'AT4', 'AD5', 'FD6', 'CE7', 'addr', 'lng', 'lat']]
theater.head()
```

```python
show.columns
```

```python
show1 = show.copy()
show1 = show1[['place', 'theater','name','genre']]
```

```python
show1.head()
```

```python
for i in range(len(theater)):
    theater.name[i] = theater.name[i].strip()
    theater.theater_name[i] = theater.theater_name[i].strip()

theater.head()
```

```python
for i in range(len(show1)):
    show1.place[i] = show1.place[i].strip()
    show1.theater[i] = show1.theater[i].strip()

show1.head()
```

```python
show.genre.unique()
```

```python
len(show1.theater.unique())
```

```python
show1.pivot_table(index=['place','theater'], columns='genre', aggfunc='count').reset_index().shape
```

```python
show_genre = show1.pivot_table(index=['place','theater'], columns='genre', aggfunc='count', margins=True).reset_index()
show_genre.head()
```

```python
show_genre_exp = show_genre.copy()
show_genre_exp.columns = ['place','theater','korea_opera','dance','musical','compound','play','opera','classic','total']
show_genre_exp.head()
```

```python
show_genre_exp.info()
```

```python
show_genre_exp = show_genre_exp.fillna(0)
show_genre_exp.head()
```

```python
show_genre_exp.columns = ['name', 'theater_name', 'korea_opera', 'dance', 'musical', 'compound',
       'play', 'opera', 'classic', 'total']
show_genre_exp.head()
```

```python
print(theater.shape)
theater.head()
```

## theater에 join

```python
df_merge = pd.merge(theater, show_genre_exp, on=['name','theater_name'], how='left')
print(df_merge.shape)
df_merge.head()
```

## outer로 join

```python
df_merge_out = pd.merge(theater, show_genre_exp, on=['name','theater_name'], how='outer')
print(df_merge_out.shape)
df_merge_out.head()
```
